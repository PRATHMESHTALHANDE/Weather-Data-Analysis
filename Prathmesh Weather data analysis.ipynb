```python
import pandas as pd
import matplotlib.pyplot as plt

data['Date/Time'] = pd.to_datetime(data['Date/Time'])

plt.figure(figsize=(8, 6))
plt.hist(data['Rel Hum_%'], bins=20, edgecolor='black')
plt.title('Humidity Distribution')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Frequency')

frequency, bins, _ = plt.hist(data['Rel Hum_%'], bins=20, edgecolor='black', alpha=0)
max_freq_index = frequency.argmax()  
max_freq_bin = bins[max_freq_index]  

date_max_frequency = data[(data['Rel Hum_%'] >= max_freq_bin) & (data['Rel Hum_%'] <= bins[max_freq_index + 1])]['Date/Time'].mode()[0]

plt.annotate(f'Date: {date_max_frequency.strftime("%Y-%m-%d %H:%M")}', xy=(max_freq_bin, max(frequency)),
             xytext=(max_freq_bin + 3, max(frequency) - 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10)

plt.tight_layout()
plt.show()
```

# Q17) Plot line graph count of month for whole year

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r"C:\Users\roari\Downloads\1. Weather Data.csv")

data['Date/Time'] = pd.to_datetime(data['Date/Time'])

data['Month'] = data['Date/Time'].dt.month
weather_condition = 'Mainly Clear'  

specific_weather = data[data['Weather'] == weather_condition]

max_occurrences_per_month = specific_weather.groupby('Month').size().max(level=0).reset_index(name='Occurrences')

plt.figure(figsize=(10, 6))
sns.lineplot(data=max_occurrences_per_month, x='Month', y='Occurrences', marker='o', linestyle='-')
plt.title(f'Highest Occurrences of {weather_condition} for Each Month')
plt.xlabel('Month')
plt.ylabel('Occurrences')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

# Q18) Can you plot distribution of different weather conditions for whole year in pie chart?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r"C:\Users\roari\Downloads\1. Weather Data.csv")

data['Date/Time'] = pd.to_datetime(data['Date/Time'])

data['Month'] = data['Date/Time'].dt.month

fig, axs = plt.subplots(4, 3, figsize=(15, 15))
axs = axs.flatten()

for i, month in enumerate(range(1, 13)):
 
    specific_month_data = data[data['Month'] == month]
    
    weather_counts = specific_month_data['Weather'].value_counts()
    
    labels = weather_counts.index
    sizes = weather_counts.values
    axs[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f"Weather Distribution - Month {month}")

plt.tight_layout()
plt.show()
```

# Q19) Can you plot bubble chart with different colours for each weather conditions for month for all 12 months?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r"C:\Users\roari\Downloads\1. Weather Data.csv")

data['Date/Time'] = pd.to_datetime(data['Date/Time'])

data['Month'] = data['Date/Time'].dt.month

unique_conditions = data['Weather'].unique()
num_conditions = len(unique_conditions)
colors = sns.color_palette("husl", num_conditions)

fig, axs = plt.subplots(4, 3, figsize=(15, 15))
axs = axs.flatten()

for i, month in enumerate(range(1, 13)):
    
    specific_month_data = data[data['Month'] == month]
    
    color_dict = {condition: colors[j] for j, condition in enumerate(unique_conditions)}
    specific_month_data['Color'] = specific_month_data['Weather'].map(color_dict)
    
    sns.scatterplot(ax=axs[i], data=specific_month_data, x='Month', y='Weather', size='Visibility_km',
                    sizes=(30, 500), hue='Weather', palette=color_dict, legend=False)
    axs[i].set_title(f"Weather Distribution - Month {month}")
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('Weather')
    axs[i].set_yticklabels(axs[i].get_yticks(), rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r"C:\Users\roari\Downloads\1. Weather Data.csv")

data['Date/Time'] = pd.to_datetime(data['Date/Time'])

data['Month'] = data['Date/Time'].dt.month

unique_conditions = data['Weather'].unique()
num_conditions = len(unique_conditions)
colors = sns.color_palette("husl", num_conditions)

fig, axs = plt.subplots(4, 3, figsize=(15, 15))
axs = axs.flatten()

for i, month in enumerate(range(1, 13)):
   
    specific_month_data = data[data['Month'] == month]
    
    color_dict = {condition: colors[j] for j, condition in enumerate(unique_conditions)}
    specific_month_data['Color'] = specific_month_data['Weather'].map(color_dict)
    
    sns.scatterplot(ax=axs[i], data=specific_month_data, x='Month', y='Weather', size='Visibility_km',
                    sizes=(30, 500), hue='Weather', palette=color_dict, legend=False)
    axs[i].set_title(f"Weather Distribution - Month {month}")
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('Weather')
    axs[i].set_yticklabels(axs[i].get_yticks(), rotation=45)

handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=condition)
           for condition, color in zip(unique_conditions, colors)]
plt.legend(handles=handles, title='Weather Condition', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
